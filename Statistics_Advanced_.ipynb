  # Q1 What is a random variable in probability theory ?

<br>

  Ans.  A random variable is a variable whose value depends on the outcome of a random event.  
It assigns a numerical value to each possible outcome.

Types:
- **Discrete random variable:** Takes specific values (e.g., number of heads in 3 coin tosses).
- **Continuous random variable:** Takes any value in a range (e.g., height of students).

**Example:**  
Rolling a die — the random variable \(X\) can take values from 1 to 6 depending on the outcome.

# Q2 What are the types of random variables ?

Ans.  Random variables are classified into two main types:

1. **Discrete Random Variable**  
   - Takes a finite or countable number of possible values.  
   - Usually comes from counting outcomes.  
   - **Example:** Number of heads in 3 coin tosses (possible values: 0, 1, 2, 3).

2. **Continuous Random Variable**  
   - Can take any value within a given range.  
   - Usually comes from measuring something.  
   - **Example:** Height of students in a class (can be any value like 160.2 cm, 170.5 cm, etc.).

These types help in choosing the correct probability distribution for analysis.

#Q3 Explain the difference between discrete and continuous distributions ?

Ans. A probability distribution describes how probabilities are assigned to values of a random variable.  
It can be **discrete** or **continuous** based on the type of variable.

1. **Discrete Distribution**  
   - Applies to discrete random variables (countable values).  
   - Probability is assigned to each individual value.  
   - **Example:** Probability of getting 0, 1, 2, or 3 heads when tossing 3 coins.

2. **Continuous Distribution**  
   - Applies to continuous random variables (uncountable values within a range).  
   - Probability is described over an interval using a probability density function (PDF).  
   - **Example:** Probability of a person’s height being between 160 cm and 170 cm.

**Key Difference:**  
In discrete distributions, probabilities are for exact values.  
In continuous distributions, probabilities are for ranges of values.

#Q4 What is a binomial distribution, and how is it used in probability ?

Ans. A **binomial distribution** is a probability distribution that describes the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes: **success** or **failure**.

**Characteristics:**
- There are a fixed number of trials (**n**).
- Each trial has two outcomes (success or failure).
- The probability of success (**p**) is the same for each trial.
- The trials are independent.

**Formula:**
P(X = k) = C(n, k) * (p^k) * (1 - p)^(n - k)  
where:
- *n* = total trials  
- *k* = number of successes  
- *p* = probability of success  
- C(n, k) = n! / [k!(n - k)!]

**Example:**
If a coin is tossed 5 times, the binomial distribution can give the probability of getting exactly 3 heads.

**Uses:**
- Quality control (number of defective products in a batch)
- Medical studies (number of patients responding to treatment)
- Marketing (number of people buying a product after seeing an ad)

#Q5 What is the standard normal distribution, and why is it important ?

<br>

Ans.  The **standard normal distribution** is a special case of the normal (bell-shaped) distribution where:
- The mean (**μ**) = 0
- The standard deviation (**σ**) = 1

It is a symmetric curve centered at zero, and the total area under the curve equals 1.

**Importance:**
- It allows us to compare different normal distributions by converting data values into **z-scores**.
- Many statistical tests and confidence intervals are based on the standard normal distribution.
- The probabilities for different ranges can be found using standard normal tables or statistical software.

**Example:**
If a student’s exam score is converted to a z-score of 1.5, it means the score is 1.5 standard deviations above the mean.

#Q6 What is the Central Limit Theorem (CLT), and why is it critical in statistics ?

<br>

Ans.  The **Central Limit Theorem (CLT)** states that when we take many random samples from any population and calculate their means, the distribution of those sample means will approach a **normal distribution** as the sample size increases — regardless of the shape of the original population.

**Key Points:**
- Works best when sample size is large (n ≥ 30 is common).
- The mean of the sample means will be close to the population mean.
- The spread of the sample means is measured by the standard error.

**Why it is important:**
- It allows us to use normal distribution methods (like z-scores) even when the population is not normally distributed.
- It is the foundation for many statistical tests and confidence intervals.

**Example:**
If we measure the heights of 40 students repeatedly in different random groups, the averages from each group will form a normal distribution, even if the original heights are not perfectly normal.

#Q7 What is the significance of confidence intervals in statistical analysis ?

Ans.  A **confidence interval (CI)** is a range of values, derived from sample data, that is likely to contain the true population parameter (like mean or proportion) with a certain level of confidence.

**Key Points:**
- The confidence level (e.g., 95%) tells us how sure we are that the interval contains the true value.
- A wider interval means more uncertainty; a narrower interval means more precision.
- Confidence intervals account for sample variability and give a better picture than a single point estimate.

**Why it is significant:**
- Helps in decision-making by showing the reliability of an estimate.
- Gives both the estimate and the margin of error.
- Used in research, surveys, quality control, and experiments.

**Example:**
If the average exam score is 75 with a 95% CI of (72, 78), we are 95% confident that the true average score for the whole population lies between 72 and 78.

#Q8  What is the concept of expected value in a probability distribution ?

<br>

Ans. The **expected value (EV)** is the long-term average outcome of a random variable over many trials of an experiment.  
It is calculated by multiplying each possible value by its probability and then adding these results.

**Formula:**
E(X) = Σ [x * P(x)]  
where:
- *x* = possible value  
- *P(x)* = probability of that value  

**Key Points:**
- Represents the "average" result if the experiment is repeated many times.
- Can be used in decision-making, risk analysis, and forecasting.

**Example:**
If you roll a fair six-sided die, the expected value is:  
E(X) = (1+2+3+4+5+6) / 6 = 3.5  
This means over many rolls, the average result will be around 3.5.

####Q 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean_value = np.mean(data)
std_dev_value = np.std(data)

# Display results
print("Mean:", mean_value)
print("Standard Deviation:", std_dev_value)

# Draw histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

####Q 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.  daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
####● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.
####● Write the Python code to compute the mean sales and its confidence interval.

<br>


Ans.  **Explanation:**
The **Central Limit Theorem (CLT)** states that the distribution of sample means will be approximately normal if we take enough random samples from a population, even if the original data is not perfectly normal.

To estimate the **average daily sales** with a **95% confidence interval**:
1. Calculate the **sample mean** (average sales from the given data).
2. Calculate the **standard error** = (sample standard deviation) / √(sample size).
3. Use the **z-score** for 95% confidence (≈ 1.96) to find the margin of error.
4. The **confidence interval** is:
   - Lower bound = mean - (z * standard error)
   - Upper bound = mean + (z * standard error)

This gives a range where we expect the true average daily sales to lie, with 95% confidence.

```python
import numpy as np
import math

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
sales_array = np.array(daily_sales)

# Calculate mean and standard deviation
mean_sales = np.mean(sales_array)
std_dev_sales = np.std(sales_array, ddof=1)  # sample standard deviation

# Sample size
n = len(sales_array)

# Standard error
standard_error = std_dev_sales / math.sqrt(n)

# Z-score for 95% confidence
z_score = 1.96

# Margin of error
margin_of_error = z_score * standard_error

# Confidence interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

# Display results
print("Mean Sales:", mean_sales)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```
